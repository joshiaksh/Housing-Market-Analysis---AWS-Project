```python

import pandas as pd
pd.options.mode.chained_assignment = None 
df =pd.read_csv("/Users/mgkulkar/OneDrive/Akshata Study Material/Study/Data Engineering/HousingMkt/NY-House-Dataset.csv")

df.head()
```

```python


HOUSE_TYPE = df[['TYPE']].drop_duplicates().reset_index(drop=True)
HOUSE_TYPE['TYPE_ID']=HOUSE_TYPE.index

HOUSE_TYPE = HOUSE_TYPE[['TYPE_ID','TYPE']]
HOUSE_TYPE

HOUSE_TYPE.to_csv("/Users/mgkulkar/OneDrive/Akshata Study Material/Study/Data Engineering/HOUSE_TYPE.csv",index=False)
```

```python
BROKER = df[['BROKERTITLE']].reset_index()
BROKER['BROKER_ID']=BROKER.index

BROKER = BROKER[['BROKER_ID','BROKERTITLE']]
BROKER

BROKER.to_csv("/Users/mgkulkar/OneDrive/Akshata Study Material/Study/Data Engineering/BROKER.csv",index=False)
```

```python
HOUSE_ADDRESS = df[['ADDRESS','STATE','MAIN_ADDRESS','ADMINISTRATIVE_AREA_LEVEL_2']].reset_index()
HOUSE_ADDRESS['ADDRESS_ID']=HOUSE_ADDRESS.index

HOUSE_ADDRESS = HOUSE_ADDRESS[['ADDRESS_ID','ADDRESS','STATE','MAIN_ADDRESS','ADMINISTRATIVE_AREA_LEVEL_2']]
HOUSE_ADDRESS

HOUSE_ADDRESS.to_csv("/Users/mgkulkar/OneDrive/Akshata Study Material/Study/Data Engineering/HOUSE_ADDRESS.csv",index=False)
```

```python
LOCALITY = df[['LOCALITY','SUBLOCALITY','STREET_NAME','LONG_NAME','FORMATTED_ADDRESS']].reset_index()
LOCALITY['LOCALITY_ID']=LOCALITY.index

LOCALITY = LOCALITY[['LOCALITY_ID','LOCALITY','SUBLOCALITY','STREET_NAME','LONG_NAME','FORMATTED_ADDRESS']]
LOCALITY

LOCALITY.to_csv("/Users/mgkulkar/OneDrive/Akshata Study Material/Study/Data Engineering/LOCALITY.csv",index=False)
```

```python

HOUSE_LOCATION = df[['LATITUDE','LONGITUDE']].reset_index(drop=True)
HOUSE_LOCATION['LOCATION_ID']=HOUSE_LOCATION.index

HOUSE_LOCATION = HOUSE_LOCATION[['LOCATION_ID','LATITUDE','LONGITUDE']]
HOUSE_LOCATION

HOUSE_LOCATION.to_csv("/Users/mgkulkar/OneDrive/Akshata Study Material/Study/Data Engineering/HOUSE_LOCATION.csv",index=False)
```

```python
HOUSE = df.merge(BROKER,on = 'BROKERTITLE')\
            .merge(HOUSE_TYPE,on='TYPE')\
            .merge(HOUSE_ADDRESS,on = ['ADDRESS'])\
            .merge(LOCALITY, on =['LOCALITY'])\
            .merge(HOUSE_LOCATION,on = ['LATITUDE','LONGITUDE'])

HOUSE.drop_duplicates()

HOUSE.to_csv("/Users/mgkulkar/OneDrive/Akshata Study Material/Study/Data Engineering/HOUSE.csv",index=False) 
```
